```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import StandardScaler
from matplotlib import*
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("USA.xlsx")#Leemos el fichero xlsx
```

```python
df
```

```python
ciudades=df.iloc[:,1:8]
def replace_name(x):
    return ciudades[ciudades['ciudadesId']==x].title.values[0]
df1 = ciudades.replace(np.nan, 0, regex=True)
std = StandardScaler().fit_transform(df1)
```

```python
mean_vec = np.mean(std, axis=0)
cov_mat = (std - mean_vec).T.dot((std - mean_vec)) / (std.shape[0]-1)
print('Covariance matrix \n%s' %cov_mat)
```

```python
cov_mat = np.cov(std.T)
eig_vals, eig_vecs = np.linalg.eig(cov_mat)
print('Eigenvectors \n%s' %eig_vecs)
print('\nEigenvalues \n%s' %eig_vals)
```

```python
eig_pairs = [(np.abs(eig_vals[i]), eig_vecs[:,i]) for i in range(len(eig_vals))]
print('Eigenvalues in descending order:')
for i in eig_pairs:
    print(i[0])
```
